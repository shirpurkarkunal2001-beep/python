```python
import pandas as pd

# Load dataset
df = pd.read_csv(r"C:\Users\karan.shirpurkar\Downloads\AG Calibration Raw Data-results-2025-08-11-01-38-12\AG Calibration Raw Data-result-1-2025-08-11-01-37-48.csv")

# Ensure time format & sort by complete_time descending
df["complete_time"] = pd.to_datetime(df["complete_time"])
df = df.sort_values("complete_time", ascending=False)

# Add variation columns
df["length_variation"] = df["length"] - df["defined_l"]
df["breadth_variation"] = df["breadth"] - df["defined_b"]
df["height_variation"] = df["height"] - df["defined_h"]
df["weight_variation"] = df["weight"] - df["defined_wt"]

# Accuracy % summary
summary = (
    df.groupby(["machine_category", "machineusername", "feedlane"])
    .agg(
        total_wbn=("wbn", "count"),
        distinct_wbn=("wbn", "nunique"),
        dim_accuracy_percent=("dim_acc", lambda x: 100 * x.mean()),
        weight_accuracy_percent=("weight_acc", lambda x: 100 * x.mean()),
        total_accuracy_percent=("total_acc", lambda x: 100 * x.mean()),
    )
    .reset_index()
)

# Variation summary (mean variation per machine/feedlane)
variation_summary = (
    df.groupby(["machine_category", "machineusername", "feedlane"])
    .agg(
        length_variations=("length_variation", "mean"),
        breadth_variations=("breadth_variation", "mean"),
        height_variations=("height_variation", "mean"),
        weight_variations=("weight_variation", "mean"),
    )
    .reset_index()
)

# Merge accuracy & variation summaries
final_summary = pd.merge(summary, variation_summary, on=["machine_category", "machineusername", "feedlane"])

# Final dataframe is sorted by complete_time already
import caas_jupyter_tools
caas_jupyter_tools.display_dataframe_to_user("Machine Feedlane Summary", final_summary)
```
